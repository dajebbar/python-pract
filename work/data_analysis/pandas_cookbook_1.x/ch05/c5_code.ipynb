 By default, `.value_counts` does not show missing values, but
we can use the `dropna` parameter to fix that

```python
fueleco.drive.value_counts(dropna=False)
```

the *rangeA* column has an object data type. However, if we use categorical method, `.value_counts`, to examine it, we see that it is not really categorical, but a numeric column posing as a category.

```python
fueleco.rangeA.value_counts(dropna=False)
```

Another way to find offending characters is to use the `.str.extract` method with a regular expression:

```python
(
    fueleco.rangeA.str.extract(r'([^0-9.])')
    .dropna()
    .apply(lambda row: ''.join(row), axis=1)
    .value_counts()
)
```

```python
 set(fueleco.rangeA.apply(type))
```

Here is the count of missing values:

```python
fueleco.rangeA.isna().sum()
```

According to the fueleconomy.gov website, the rangeA value represents the range for the second fuel type of dual fuel vehicles (E85, electricity, CNG, and LPG). Using pandas, we can replace the missing values with zero, replace dashes with slashes, then split and take the mean value of each row (in the case of a dash/slash):

```python
(
    fueleco.rangeA.fillna('0')
    .str.replace('-', '/')
    .str.split('/', expand=True)
    .astype(float)
    .mean(axis=1)
)
```

We can also treat numeric columns as categories by binning them. There are two powerful functions in pandas to aid binning, cut and qcut. We can use cut to cut into equal-width bins, or bin widths that we specify. For the rangeA column, most of the values were empty and we replaced them with 0, so 10 equal-width bins look like this:

```python
(
    fueleco.rangeA.fillna('0')
    .str.replace('-', '/')
    .str.split('/', expand=True)
    .astype(float)
    .mean(axis=1)
    .pipe(lambda ser_ : pd.cut(ser_, 10))
    .value_counts()
)
```

If you are concerned with whether the data is normal, you can quantify this with numbers and visualizations using the SciPy library.
The **Kolmogorov-Smirnov** test can evaluate whether a distribution is normal. It provides us with a *p-value*. If this value is significant (< 0.05), then the data is not normal:

```python
from scipy import stats
```

```python
stats.kstest(fueleco.city08, cdf='norm')
```

We can plot a probability plot to see whether the values are normal. If the samples track the
line, then the data is normal

```python
fig, ax = plt.subplots(figsize=(10, 8))
stats.probplot(fueleco.city08, plot=ax)
```

```python
fig.savefig("c5-conprob.png", dpi=300)
```

## Comparing continuous values across categories
will look at mileage numbers in
different brands: Ford, Honda, Tesla, and BMW

Make a mask for the brands we want and then use a group by operation to look at
the mean and standard deviation for the city08 column for each group of cars

```python
brands = ["Ford", "Honda", "Tesla", "BMW"]
mask = fueleco.make.isin(brands)

fueleco[mask].groupby(['make'])['city08'].agg(['mean', 'std'])
```

Visualize the city08 values for each make with seaborn

```python
g = sns.catplot(
    data=fueleco[mask],
    x='make',
    y='city08',
    kind='box',
)
```

```python
 g.ax.figure.savefig("c5-catbox.png", dpi=300)
```

If the summary statistics change for the different makes, that is a strong indicator that the makes have different characteristics. The central tendency (mean or median) and the variance (or standard deviation) are good measures to compare. We can see that Honda gets better city mileage than both BMW and Ford but has more variance, while Tesla is better than all of them and has the tightest variance.

One drawback of a boxplot is that while it indicates the spread of the data, it does not reveal how many samples are in each make. You might naively think that each boxplot has the same number of samples. We can quantify that this is not the case with pandas

```python
fueleco[mask].groupby('make').city08.count()
```

Another option is to do a swarm plot on top of the box plots:

```python
g = sns.catplot(
    data=fueleco[mask],
    x='make',
    y='city08',
    kind='box',
)
```

```python
sns.swarmplot(
    x="make",
    y="city08",
    data=fueleco[mask],
    color="k",
    size=1,
    ax=g.ax,
)
```

```python
 g.ax.figure.savefig("c5-catbox2.png", dpi=300)
```

break each of new plots into its own
graph by using the **col** parameter

```python
g = sns.catplot(
    data=fueleco[mask],
    x='make',
    y='city08',
    kind='box',
    col="year",
    col_order=list(range(2012, 2019)),
    col_wrap=2,
)
```

```python
 g.axes[0].figure.savefig("c5-catboxcol.png", dpi=300)
```

embed the new dimension in the same plot by using the hue parameter

```python
g = sns.catplot(
    data=fueleco[mask],
    x='make',
    y='city08',
    kind='box',
    hue="year",
    hue_order=list(range(2012, 2019)),
)
```

```python
g.ax.figure.savefig("c5-catboxhue.png", dpi=300)
```
