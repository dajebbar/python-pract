## Summary statistics
Summary statistics include the mean, quartiles, and standard deviation. The `.describe`
method will calculate these measures on all of the numeric columns in a DataFrame

```python
import numpy as np
import pandas as pd
```

```python
fueleco = pd.read_csv("./vehicles.csv.zip")
fueleco.head()
```

Call individual summary statistics methods such as `.mean`, `.std`, and `.quantile`

```python
fueleco.mean()
```

```python
fueleco.std()
```

```python
fueleco.quantile(q=np.linspace(0, 1, 5)).T
```

Call the `.describe` method

```python
fueleco.describe(percentiles=np.linspace(0, 1, 4, endpoint=False)).T
```

To get summary statistics on the object columns, use the `.include` parameter

```python
fueleco.describe(include='object').T
```
