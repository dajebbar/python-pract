This is the process of sifting through the data and trying to make sense of the individual columns and the relationships between them.

We are going to use a dataset from www.fueleconomy.gov that provides information about
makes and models of cars from 1984 through 2018. Using EDA we will explore many of the
columns and relationships found in this data.

## Summary statistics
Summary statistics include the mean, quartiles, and standard deviation. The `.describe`
method will calculate these measures on all of the numeric columns in a DataFrame

```python
import numpy as np
import pandas as pd
```

```python
fueleco = pd.read_csv("./vehicles.csv.zip")
fueleco.head()
```

Call individual summary statistics methods such as `.mean`, `.std`, and `.quantile`

```python
fueleco.mean()
```

```python
fueleco.std()
```

```python
fueleco.quantile(q=np.linspace(0, 1, 5)).T
```

Call the `.describe` method

```python
fueleco.describe(percentiles=np.linspace(0, 1, 4, endpoint=False)).T
```

To get summary statistics on the object columns, use the `.include` parameter

```python
fueleco.describe(include='object').T
```

## Column types
Explore the column types.

```python
# Inspect the .dtypes attribute
fueleco.dtypes
```

Summarize the types of columns

```python
fueleco.dtypes.value_counts()
```

Object types may be strings or categorical data, but they could also be numeric-like values that need to be nudged a little so that they are numeric. 
I typically leave integer columns alone. 
I tend to treat them as continuous values. If the values are float values, this indicates that the column could be:
-  All floating-point values with no missing values
-  Floating-point values with missing values
-  Integer values that were missing some values and hence converted to floats

Save some memory by converting these columns to columns that use less memory.

```python
fueleco.select_dtypes(include='int64').describe().T
```

We can see that the city08 and comb08 columns don't go above 150.

```python
np.iinfo(np.int8)
```

```python
np.iinfo(np.int16)
```

```python
fueleco[['city08', 'comb08']].info(memory_usage='deep')
```

```python
(
    fueleco[['city08', 'comb08']]
    .assign(
        city08 = fueleco.city08.astype(np.int16),
        comb08 = fueleco.comb08.astype(np.int16),
    )
    .info(memory_usage='deep')
)
```

```python
fueleco[['make', 'model']].nunique()
```

```python
fueleco[['make', 'model']].info(memory_usage='deep')
```

```python
fueleco[['make']].info(memory_usage='deep')
```

```python
(
    fueleco[['make']]
    .assign(
        make = fueleco.make.astype('category')
    )
    .info(memory_usage='deep')
)
```

```python
fueleco[['model']].info(memory_usage='deep')
```

```python
(
    fueleco[['model']]
    .assign(
        model = fueleco.model.astype('category')
    )
    .info(memory_usage='deep')
)
```
