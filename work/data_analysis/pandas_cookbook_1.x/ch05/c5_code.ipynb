 By default, `.value_counts` does not show missing values, but

the *rangeA* column has an object data type. However, if we use categorical method, `.value_counts`, to examine it, we see that it is not really categorical, but a numeric column posing as a category.

```python
fueleco.rangeA.value_counts(dropna=False)
```

Another way to find offending characters is to use the `.str.extract` method with a regular expression:

```python
(
    fueleco.rangeA.str.extract(r'([^0-9.])')
    .dropna()
    .apply(lambda row: ''.join(row), axis=1)
    .value_counts()
)
```

```python
 set(fueleco.rangeA.apply(type))
```

Here is the count of missing values:

```python
fueleco.rangeA.isna().sum()
```

According to the fueleconomy.gov website, the rangeA value represents the range for the second fuel type of dual fuel vehicles (E85, electricity, CNG, and LPG). Using pandas, we can replace the missing values with zero, replace dashes with slashes, then split and take the mean value of each row (in the case of a dash/slash):

```python
(
    fueleco.rangeA.fillna('0')
    .str.replace('-', '/')
    .str.split('/', expand=True)
    .astype(float)
    .mean(axis=1)
)
```

We can also treat numeric columns as categories by binning them. There are two powerful functions in pandas to aid binning, cut and qcut. We can use cut to cut into equal-width bins, or bin widths that we specify. For the rangeA column, most of the values were empty and we replaced them with 0, so 10 equal-width bins look like this:

```python
(
    fueleco.rangeA.fillna('0')
    .str.replace('-', '/')
    .str.split('/', expand=True)
    .astype(float)
    .mean(axis=1)
    .pipe(lambda ser_ : pd.cut(ser_, 10))
    .value_counts()
)
```

Pearson correlation is intended to show the strength of a linear relationship. If the two continuous columns do not have a linear relationship, another option is to use Spearman correlation. ## Comparing categorical values with categorical values
In this section, we will focus on dealing with multiple categorical values. One thing to keep in mind is that continuous columns can be converted into categorical columns by binning the values.

look at makes and vehicle class.

```python
fueleco[['make', 'VClass']].head()
# fueleco.select_dtypes(include='object').columns
```

```python
fueleco.VClass.nunique()
```

Lower the cardinality. Limit the VClass column to six values, in a simple class column, SClass. Only use Ford, Tesla, BMW, and Toyota:

```python
def generalize(ser, match_name, default):
  seen = None
  for match, name in match_name:
    mask = ser.str.contains(match)
    if seen is None:
      seen = mask
    else:
      seen |= mask
    ser = ser.where(~mask, name)
  ser = ser.where(seen, default)
  return ser
```

```python
makes = ["Ford", "Tesla", "BMW", "Toyota"]
data = fueleco[fueleco.make.isin(makes)].assign(
    SClass = lambda df_: generalize(
        df_.VClass,
        [
         ("Seaters", "Car"),
         ("Car", "Car"),
         ("Utility", "SUV"),
         ("Truck", "Truck"),
         ("Van", "Van"),
         ("van", "Van"),
         ("Wagon", "Wagon"),
         ],
         default='Other'
    )
)
```

Summarize the counts of vehicle classes for each make:

```python
data.groupby(["make", "SClass"]).size().unstack()
```

Use the crosstab function instead of the chain of pandas commands

```python
pd.crosstab(data.make, data.SClass)
```

Add more dimensions:

```python
pd.crosstab([data.year, data.make], [data.SClass, data.VClass])
```

Use CramÃ©r's V measure  to indicate the categorical correlation

```python
import scipy.stats as ss
```

```python
def cramers_v(x, y):
  confusion_matrix = pd.crosstab(x, y)
  chi2 = ss.chi2_contingency(confusion_matrix)[0]
  n = confusion_matrix.sum().sum()
  phi2 = chi2 / n
  r, k = confusion_matrix.shape
  phi2corr = max(
      0, phi2 - ((k - 1) * (r - 1)) / (n - 1)
  )
  rcorr = r - ((r - 1) ** 2) / (n - 1)
  kcorr = k - ((k - 1) ** 2) / (n - 1)
  return np.sqrt(
      phi2corr / min((kcorr - 1), (rcorr - 1))
  )
```

```python
cramers_v(data.make, data.SClass)
```

The .corr method accepts a callable as well, so an alternative way to invoke this is the following:

```python
data.make.corr(data.SClass, cramers_v)
```

Visualize the cross tabulation as a bar plot:

```python
fig, ax = plt.subplots(figsize=(10, 8))
(
    data.pipe(
        lambda df_: pd.crosstab(df_.make, df_.SClass)
        
    ).plot.bar(ax=ax)
)
```

Visualize the relative sizes of the groups by normalizing the cross tabulation and
making a stacked bar chart:
